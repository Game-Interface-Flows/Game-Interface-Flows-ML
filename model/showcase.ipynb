```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
RANDOM_SEEED = 100
random.seed(RANDOM_SEEED)
```

```python
def display_random_images(path: str, num_images: int = 9) -> None:
    if not os.path.exists(path):
        print(f"The directory {path} does not exist.")
        return

    images = [
        os.path.join(path, img)
        for img in os.listdir(path)
        if img.lower().endswith((".png", ".jpg", ".jpeg"))
    ]

    if len(images) > num_images:
        images = random.sample(images, num_images)
    else:
        print(
            f"Not enough images in {path} to sample {num_images}. Showing all available images."
        )

    fig, axs = plt.subplots(
        nrows=3, ncols=3, figsize=(16, 9)
    )
    axs = axs.ravel()

    title_font = {
        "family": "Times New Roman",
        "color": "#000",
        "weight": "normal",
        "size": 14,
    }

    for i, img_path in enumerate(images):
        img = Image.open(img_path)
        axs[i].imshow(img)
        axs[i].set_title(os.path.basename(img_path), fontdict=title_font)
        axs[i].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
class_0_path = "data/0/"
display_random_images(class_0_path)
```

```python
class_1_path = "data/1/"
display_random_images(class_1_path)
```
